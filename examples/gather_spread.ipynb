```python
import redframes as rf
```

```python
df = rf.DataFrame({
    'foo': [1, 4], 
    'bar': [2, 5], 
    'baz': [3, 6]
})

df
```

```python
df = df.gather(["foo", "bar"], into=("variable", "value"))
df
```

```python
df.spread("variable", "value")
```

```python
df = rf.DataFrame({
    'foo': [1, 4], 
    'bar': [2, 5], 
    'baz': [3, 6]
})

df
```
